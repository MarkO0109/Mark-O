## - Purpose of this Model - 
  
This data represents the data used in the test run for this assignment.**
   
It represents the Pokemon in the tier of **NeverUsed (NU)**, the second lowest tier. The usage stats are from July 2018, so this is not the finalized version of it (as the final version is from November 2019). 
    
**Smogon**, although unofficial, is one of the most popular metagame formats of **Pokemon**, which was established in December 2004. Its goal is to balance **Pokemon** for competitive play by restrictions (no same **Pokemon** in teams, Pokemon in higher tiers cannot be used in lower tiers, **Pokemon** must not use accuracy evasion moves/items, etc). In this case, I'll focus on the **NU tier**. 
   
**NU** allows Pokemon from **NU**, and also **PUBL** and **PU**. Pokemon in **PUBL** are Pokemon that got banned from **PU**, but are not in **NU** by usage. In order to be **NU**, the usage percentage has to be **3.41%**. If not, then the Pokemon will drop to **PUBL** or **PU**. 
   
The **Smogon NU tier link** is here: https://www.smogon.com/dex/sm/formats/nu/
   
The **Smogon PUBL tier link** is here: https://www.smogon.com/dex/sm/formats/publ/
   
The **Smogon PU tier link** is here: https://www.smogon.com/dex/sm/formats/pu/

It fits well with teams due to its good defensive typing, ability and support such as **Light Screen** and **Reflect**.

With it also being heavy, it can make use of STAB **Heavy Slam**.
